# NORMALIZACIÓN

```python
def normalizacionColumna(df, i):
    columns = df.columns.values
    df[columns[i]] = (df[columns[i]] - data[columns[i]].min()) / (df[columns[i]].max() - data[columns[i]].min())
```

```python
def normalizarDataset(data, indices):
    df = data.copy()
    for i in indices:
        normalizacionColumna(df, i)
    return df
```

# WEIGHTED AVERAGE

```python
def weightedAverage(data, w, l):
    df = data.copy()
    columns = df.columns.values
    wa = np.zeros(df.shape[0])
    for i in range(len(l)):
        wa += (w[i] * df[columns[l[i]]]) / sum(w)
    df['wa'] = wa
    df = df.sort_values(by=['wa'], ascending=False)
    return df
```

# MAXIMIN 

```python
def maximin(data, l):
    df = data.copy()
    columns = df.columns.values
    t = df.shape[0]
    mn = np.zeros(df.shape[0])
    for i in range(t):
        mn[i] = df[columns[l[0]]][i]
        for j in range(1,len(l)):
            if mn[i] > df[columns[l[j]]][i]:                
                mn[i] = df[columns[l[j]]][i]
    df['minVal'] = mn
    #print(mn)
    df = df.sort_values(by=['minVal'], ascending=False)
    return df
```

# LEXIMIN

```python
def leximin(data, l):
    df = data.copy()
    columns = df.columns.values
    t = df.shape[0]
    lex = [np.zeros(df.shape[0]) for i in range(len(l))]
    a = [[]  for i in range(len(l))]
    for i in range(t):
        for j in range(len(l)):
            a[j] = df[columns[l[j]]][i]
        a.sort()
        for j in range(len(l)):
            lex[j][i] = a[j]
    for j in range(len(l)):
        df['c' + str(j)] = lex[j]
    c = ['c' + str(i) for i in range(len(l))]
    df = df.sort_values(by=c, ascending=False)
    return df
```

# SKYLINES

```python
def ParetoDomina(a,b):
    mi = len([1 for i in range(len(a)) if a[i] >= b[i]])
    my = len([1 for i in range(len(a)) if a[i] > b[i]])
    if mi == len(a):
        if my > 0:
            return True
    return False
```

```python
def skylines(data, l):
    df = data.copy()
    columns = df.columns.values
    t = df.shape[0]
    for i in range(t):
        if i in df.index:
            a = [0] * len(l)
            for j in range(i + 1, t):
                if j in df.index:
                    b = [0] * len(l)
                    for k in range(len(l)):
                        a[k] = df[columns[l[k]]][i]
                        b[k] = df[columns[l[k]]][j]
                    if ParetoDomina(a,b):
                        df = df.drop(j)
                    elif ParetoDomina(b,a):
                        df = df.drop(i)
                        break
    return df
```

# KMeans
## Función

```python
def KMeans(df,k,max_it):
    columns = df.columns.values
    n,cols = df.shape
    centers = [[0]*cols for _ in range(k)]
    repeat = False
    rc = []
    it = 0
    while len(rc) != k:
        r = rnd.randint(0,n-1)
        if r not in rc:
            rc.append(r)
    for i in range(k):
        for j in range(cols):
            centers[i][j]=df[columns[j]][rc[i]]
    
    while not repeat and it<max_it:
        dist = [[0]*n for i in range(k)]
        G = [0 for i in range(n)]
        for point in range(n):
            c = 0
            for cent in range(k):
                suma = 0
                for col in range(cols):
                    suma += pow(centers[cent][col]-df[columns[col]][point],2)
                dist[cent][point] = math.sqrt(suma)
                if dist[cent][point] < dist[c][point]:
                    c = cent
            G[point] = c
        newcenters = [[0]*cols for _ in range(k)]
        for cent in range(k):
            counter = 0
            for point in range(n):
                if G[point]==cent:
                    for col in range(cols):
                        newcenters[cent][col]+= df[columns[col]][point]
                    counter += 1
            for col in range(cols):
                newcenters[cent][col] /= counter
        it+=1
        repeat = True
        for col in range(cols):
            if newcenters[:][col]!=centers[:][col]:
                repeat = False
        centers = newcenters
        
    return G,centers,it
```

# K-Nearest Neighbor

## Función

```python
def KNearestNeighbor(training_data,cluster,test_data,indices,n_neighbors):
    columns = training_data.columns.values
    n_training,cols = training_data.shape
    n_test,_ = test_data.shape
    G = [0 for i in range(n_test)]
    for dtest in range(n_test):
        distances = []
        for dtrain in range(n_training):
            suma = 0
            for col in indices:
                suma += pow(training_data[columns[col]][dtrain]-test_data[columns[col]][dtest],2)
            distances.append((cluster[dtrain],math.sqrt(suma)))
        distances.sort(key=operator.itemgetter(1))
        nearest = []
        for x in range(n_neighbors):
            nearest.append(distances[x][0])
        groupcount = [nearest.count(x) for x in range(len(indices))]
        G[dtest]=groupcount.index(max(groupcount))
    
    return G
```
